## 2) Convolutional neural network (CNN) architecture
--------------------------------------
Our final model has the following architecture:

$INPUT (54x54x3)$ <br>
${CONV}_1(32, S5) => RELU$ <br>
$CONV_2(48, S5) => RELU => DROPOUT => MAXPOOL_1(S2)$ <br>
${CONV}_3(64, S5) => RELU$ <br>
$CONV_4(128, S5) => RELU => DROPOUT => MAXPOOL_2(S2)$ <br> 
$CONV_5(192, S3) => RELU$ <br>
$CONV_6(256, 3) => RELU => DROPOUT => AVGPOOL_1(S2)$ <br>
$CONV_7(512 ,S3) => RELU$ <br>
$CONV_8(768, 3) => RELU => DROPOUT => AVGPOOL_2(S2)$ <br>
$FC_1(3072) => RELU => DROPOUT$ <br>
$FC_2(3072) => RELU => DROPOUT$ <br>
$OUTPUT (1x11, 1x11, 1x11, 1x2)$ <br>
 <br>
here:
* $CONV_{index}(feature \, map \,size, \,patch)$ 
* $[AVG/MAX]POOL_{index}(stride/kernel \,size)$ 
* symbol $S$ in parameters means same padding 
* all AVGPOOLS and MAXPOOLS is done with kernel 2x2 with stride 2 in both directions, with the same padding as well
* DROPOUT is 0.9 <br>
<br>

Output has four separated heads: 
* *Three heads for digits classification*. CNN classifies numbers not bigger than 999, each CNN output head for a different digit in the number which identifies values 0,1,...,10. 10 indicates that a given image doesn't contain a digit. 
* *One head for digits quantity in the number classification in the image*. There are five values 0,1,2,3,4 to classify. 0 means there isn't any number in given photo, 4 - there are more than 3 digits in the number. In the following, this head will be called Length head.
    
To get better results for the first three heads we use a different dataset (in which all photos contain a number) than for the last one. To optimize the training process we will use transfer training. Firstly we'll train CNN all graph (except number head) with the first dataset and then we will train only number head weights with the second dataset.

Additionally, we try a slightly different model architecture (B architecture) in which the length head uses it's own full connected layer (FC) (see architectures below). In this Jupyter notebook, B architecture approach CCN graph and training process code is given.

<img src="Pictures/CNN_architectures.png" title="CNN architectures">

```python
def calculate_digit_acc(pred_label, label, i):
  digit_label = argmax(label[:,i,:], axis = 1)
  acc = accuracy_score(pred_label[i], digit_label)
  return acc

def calculate_digit_acc_tf(pred_label, label, i):
  digit_label = tf.argmax(label[:,i,:], axis = 1)
  acc = tf.reduce_mean(tf.cast(tf.equal(pred_label[i], digit_label), tf.float32))
  return acc
```

CCN graph code.

```python
import tensorflow as tf

  
patch_size = 5; patch_size2 = 3; 
depth_1 = 32; depth_2 = 48;
depth_3 = 64; depth_4 = 128;
depth_5 = 192; depth_6 = 256;
depth_7 = 512; depth_8 = 768;

num_hidden1 = 3072; num_hidden2 = 3072;
num_len_labels = 3; image_size = 54
num_labels = 11; num_channels = 3
num_length_labels = 5
graph = tf.Graph()

with graph.as_default():

  # Input data.
  tf_train_dataset = tf.placeholder(
    tf.float32, shape=(None, image_size, image_size, num_channels), name = 'X') 
  tf_train_labels = tf.placeholder(tf.float32, shape=(None, num_len_labels, num_labels), name = 'labels')
  tf_train_length = tf.placeholder(tf.float32, shape=(None, num_length_labels), name = 'length_labels')

  tf_validate_dataset = tf.placeholder(
     tf.float32, shape=(None, image_size, image_size, num_channels), name = "val_X") 
  tf_validate_labels = tf.placeholder(tf.float32, shape=(None, num_len_labels, num_labels), name = "val_labels")
  tf_validate_length_labels = tf.placeholder(tf.float32, shape=(None, num_length_labels), name = "val_length_labels")
  
  # Define variables
  def weight_variable(shape, name):
    return tf.Variable(tf.truncated_normal(shape=shape, stddev=0.1), name=name)
  def bias_variable(shape, name):
    return tf.Variable(tf.constant(1.0, shape=shape), name=name)
  def output_weight_variable(shape, name):
    return tf.get_variable(name = name , shape=shape, initializer=tf.contrib.layers.xavier_initializer())
  def softmax_argmax(data):
    return tf.argmax(tf.nn.softmax(data), axis = 1)

  weights = {
      "layer1": weight_variable([patch_size, patch_size, num_channels, depth_1], "W1"),
      "layer2": weight_variable([patch_size, patch_size, depth_1, depth_2], "W2"),
      "layer3": weight_variable([patch_size, patch_size, depth_2, depth_3], "W3"),
      "layer4": output_weight_variable([patch_size, patch_size, depth_3, depth_4], "W4"),
      "layer5": output_weight_variable([patch_size2, patch_size2, depth_4, depth_5], "W5"),
      "layer6": output_weight_variable([patch_size2, patch_size2, depth_5, depth_6], "W6"),
      "layer7": output_weight_variable([patch_size2, patch_size2, depth_6, depth_7], "W7"),
      "layer8": output_weight_variable([patch_size2, patch_size2, depth_7, depth_8], "W8"),
      "fc_layer1": output_weight_variable([2 * 2 * depth_8, num_hidden1], "fc_W1"),
      "fc_layer2": output_weight_variable([num_hidden1, num_hidden2], "fc_W2"),
      "fc_layer3": output_weight_variable([2 * 2 * depth_8, num_hidden1], "fc_W3"),
      "fc_layer4": output_weight_variable([num_hidden1, num_hidden2], "fc_W4"),
      "output1": output_weight_variable([num_hidden2, num_labels], "Wo1"),
      "output2": output_weight_variable([num_hidden2, num_labels], "Wo2"),
      "output3": output_weight_variable([num_hidden2, num_labels], "Wo3"),
      "output4": output_weight_variable([num_hidden2, num_length_labels], "Wo4")
  
  }

  biases = {
      "layer1": tf.Variable(tf.zeros([depth_1]), name = 'b1'),
      "layer2": bias_variable([depth_2], name="b2"),
      "layer3": bias_variable([depth_3], name="b3"),
      "layer4": bias_variable([depth_4], name="b4"),
      "layer5": bias_variable([depth_5], name="b5"),
      "layer6": bias_variable([depth_6], name="b6"),
      "layer7": bias_variable([depth_7], name="b7"),
      "layer8": bias_variable([depth_8], name="b8"),
      "fc_layer1": bias_variable([num_hidden1], name="fc_b1"),
      "fc_layer2": bias_variable([num_hidden2], name="fc_b2"),
      "fc_layer3": bias_variable([num_hidden1], name="fc_b3"),
      "fc_layer4": bias_variable([num_hidden2], name="fc_b4"),
      "output1": bias_variable([num_labels], name="bo1"),
      "output2": bias_variable([num_labels], name="bo2"),
      "output3": bias_variable([num_labels], name="bo3"),
      "output4": bias_variable([num_length_labels], name="bo4"),
    }


  def cnn_layer(data, index1, index2, dropout_prob, con_pan = 'VALID', maxpool = True):           
    with tf.variable_scope("conv"+index1):
      conv1 = tf.nn.conv2d(data, weights["layer"+index1], [1, 1, 1, 1], padding="SAME", name = "con"+index1)
      relu1 = tf.nn.relu(conv1 + biases["layer"+index1], name = "relu"+index1)
      conv2 = tf.nn.conv2d(relu1, weights["layer"+index2], [1, 1, 1, 1], padding=con_pan, name = "con"+index2) 
      relu2 = tf.nn.relu(conv2 + biases["layer"+index2], name = "relu"+index2)
      dropout = tf.nn.dropout(relu2, dropout_prob, name = "drop"+index1)
      if maxpool:
        pool = tf.nn.max_pool(dropout, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME', name = 'pool'+index1)
      else:
        pool = tf.nn.avg_pool(dropout, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME', name = 'pool'+index1)
     # norm = tf.nn.local_response_normalization(pool)
    
    return pool
  
    
  # Model.
  def model(data, dropout_prob = 0.9):
    #Conv layers 
    with tf.variable_scope("CNN"):
      conv_1 = cnn_layer(data, "1","2", dropout_prob, con_pan = "SAME")
      conv_2 = cnn_layer(conv_1, "3","4", dropout_prob, con_pan = "SAME")
      conv_3 = cnn_layer(conv_2, "5","6", dropout_prob, con_pan = "VALID", maxpool = False)
      conv_4 = cnn_layer(conv_3, "7","8", dropout_prob, con_pan = "VALID", maxpool = False)
      shape = conv_4.get_shape().as_list()  
      reshape = tf.reshape(conv_4, [-1, shape[1] * shape[2] * shape[3]])
    #Full connected NN layers
    with tf.variable_scope("DIGITS_FC"):
      #digits fc nn layers
      nn_hidden_1 = tf.nn.relu(tf.matmul(reshape, weights["fc_layer1"]) + biases["fc_layer1"], name = "nn1")
      dropout_1 = tf.nn.dropout(nn_hidden_1, dropout_prob, name = "drop_nn_1")   
      nn_hidden_2 = tf.nn.relu(tf.matmul(dropout_1, weights["fc_layer2"]) + biases["fc_layer2"], name = "nn2")
      dropout_2 = tf.nn.dropout(nn_hidden_2, dropout_prob, name = "drop_nn_2")
      #Output
    with tf.variable_scope("DIGITS"):
      logits1 = tf.add(tf.matmul(dropout_2, weights["output1"]), biases["output1"], name= "L1")
      logits2 = tf.add(tf.matmul(dropout_2, weights["output2"]), biases["output2"], name= "L2")
      logits3 = tf.add(tf.matmul(dropout_2, weights["output3"]), biases["output3"], name= "L3")
    with tf.variable_scope("LENGTH_FC"):
      #number length fc nn layers
      nn_hidden_3 = tf.nn.relu(tf.matmul(reshape, weights["fc_layer3"]) + biases["fc_layer3"], name = "nn3")
      dropout_3= tf.nn.dropout(nn_hidden_3, dropout_prob, name = "drop_nn_3")   
      nn_hidden_4= tf.nn.relu(tf.matmul(dropout_3, weights["fc_layer4"]) + biases["fc_layer4"], name = "nn4")
      dropout_4 = tf.nn.dropout(nn_hidden_4, dropout_prob, name = "drop_nn_4")
       #Output
    with tf.variable_scope("LENGTH"):
      length = tf.add(tf.matmul(dropout_4, weights["output4"]), biases["output4"], name= "length")
    
    return [logits1, logits2, logits3, length]
  
  # Training computation.
  [logits1, logits2, logits3, length] = model(tf_train_dataset)
    

  logit_loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits1, labels= tf_train_labels[:,0,:]))+\
        tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits2, labels=tf_train_labels[:,1,:])) +\
        tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits3, labels=tf_train_labels[:,2,:])) 

  length_loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=length, labels= tf_train_length))
       
  # Optimizer.
  vars = tf.trainable_variables()
  train_vars1 =[]
  train_vars2 =[]
  filter_vars = ['fc_W3:0', 'fc_W4:0', 'fc_b3:0', 'fc_b4:0', 'bo4:0', 'Wo4:0']
  #filter_vars = ['bo4:0', 'Wo4:0']
  for var in vars:
    for filter_var in filter_vars:
        if filter_var in var.name:
            train_vars1.append(var)
  train_vars2 = [var for var in vars if var not in train_vars1]      
    
    
  global_step = tf.Variable(0, trainable = False)
  learning_rate = tf.train.exponential_decay(0.005, global_step, 5000, 0.95)
  optimizer = tf.train.AdagradOptimizer(learning_rate).minimize(logit_loss, global_step=global_step, var_list=train_vars2) 
  length_optimizer = tf.train.AdagradOptimizer(learning_rate).minimize(length_loss, global_step=global_step, var_list=train_vars1)
  #
  #optimizer = tf.train.AdagradOptimizer(learning_rate).minimize(loss)       
  
  tf.summary.scalar("loss", logit_loss) 
  tf.summary.scalar("length_loss", length_loss)
  tf.summary.scalar("learning_rate", learning_rate) 

  #Predictions for the training, validation, and test data.
  train_prediction = tf.stack([softmax_argmax(logits1),\
                                softmax_argmax(logits2),\
                                softmax_argmax(logits3)])

  [val_logits1, val_logits2, val_logits3, val_length] = model(tf_validate_dataset)
  validate_prediction = tf.stack([softmax_argmax(val_logits1),\
                                 softmax_argmax(val_logits2),\
                                 softmax_argmax(val_logits3)])
  validate_len_pred = softmax_argmax(val_length)
    
#   tf.summary.histogram("weigths1", weights["layer1"]); tf.summary.histogram("bias1", biases["layer1"])
#   tf.summary.histogram("weigths2", weights["layer2"]); tf.summary.histogram("bias2", biases["layer2"])
#   tf.summary.histogram("weigths3", weights["layer3"]); tf.summary.histogram("bias3", biases["layer3"])
#   tf.summary.histogram("weigths5", weights["fc_layer1"]); tf.summary.histogram("bias5", biases["fc_layer1"])
#   tf.summary.histogram("weigths6", weights["fc_layer2"]); tf.summary.histogram("bias6", biases["fc_layer2"])
   

  d1_acc = calculate_digit_acc_tf(train_prediction, tf_train_labels, 0)
  d2_acc = calculate_digit_acc_tf(train_prediction, tf_train_labels, 1)  
  d3_acc = calculate_digit_acc_tf(train_prediction, tf_train_labels, 2)  
  acc = tf.reduce_mean([d1_acc,d2_acc,d3_acc])
    
    
  tf.summary.scalar("d1_acc", d1_acc)
  tf.summary.scalar("d2_acc", d2_acc)
  tf.summary.scalar("d3_acc", d3_acc)
  tf.summary.scalar("accuracy", acc)
  

  saver = tf.train.Saver(tf.global_variables())
  merge = tf.summary.merge_all()
```

## 3) Model training

Both architectures for training has the same configuration:
* AdamOptimizer optimizer with $5x10^{-3}$ initial learning rate which decay exponentially every 5000 steps with a base of 0.96. 
* Dropouts set 0.9 
* For all classification head optimization, we use cross-entropy minimization (with softmax)
* 200 000 epochs for digits heads and 150 000 heads for length head (look above) 

Firstly, we will train digit heads with all CNN weights (except length heads and its FC layers weights, the weights variables list you can see above $train\_vars1$). 


```python
#function calculates each digit classification metrics and saves it for tensorflowboard 
def eval_valid_data(step, val_batch_size, validate_writer):
    acc1 = []; acc2 = []; acc3 = []; val_pred = []
    val_steps = len(validate_data)// val_batch_size
    for j in range(val_steps):
        
        val_offset = (j * val_batch_size) % (len(validate_data) - val_batch_size)
        val_batch_data = validate_data[val_offset:(val_offset + val_batch_size), :, :, :]
        val_batch_labels = validate_labels[val_offset:(val_offset + val_batch_size), :3]    
        
        val_batch_pred = validate_prediction.eval(feed_dict={tf_validate_dataset : val_batch_data})
        acc1.append(calculate_digit_acc(val_batch_pred, val_batch_labels, 0))
        acc2.append(calculate_digit_acc(val_batch_pred, val_batch_labels, 1))
        acc3.append(calculate_digit_acc(val_batch_pred, val_batch_labels, 2))  
        val_pred.append(np.mean([acc1,acc2,acc3]))
        
    acc1 = np.mean(acc1); acc2 = np.mean(acc2);
    acc3 = np.mean(acc3); val_acc = np.mean(val_pred)
    
    val_summary = tf.Summary()
    val_summary.value.add(tag="d1_acc", simple_value= acc1)
    val_summary.value.add(tag="d2_acc", simple_value= acc2)
    val_summary.value.add(tag="d3_acc", simple_value= acc3)
    val_summary.value.add(tag="accuracy", simple_value= val_acc)
    validate_writer.add_summary(val_summary, step)
    
    print('D1: %.3f  D2: %.3f  D3: %.3f' % (acc1, acc2, acc3))
    print('Validate acc at step %d: %f' % (step, val_acc))

    #functions calculate number length classification accuracy and saves it for tensorflowboard 
def calculate_len_digit_acc(pred_label, label):
  digit_label = argmax(label, axis = 1)
  acc = accuracy_score(pred_label, digit_label)
  return acc

def eval_length_valid_data(step, val_batch_size, validate_writer):
    acc = []; acc_background = []; acc_more_than_3 = []
    val_steps = len(validate_data)// val_batch_size
    for j in range(val_steps):
        
        val_offset = (j * val_batch_size) % (len(validate_data) - val_batch_size)
        val_batch_data = validate_data[val_offset:(val_offset + val_batch_size)]
        val_batch_labels = argmax(validate_length_labels[val_offset:(val_offset + val_batch_size)], axis = 1)     
        
        val_batch_pred = validate_len_pred.eval(feed_dict={tf_validate_dataset : val_batch_data})
        acc.append(accuracy_score(val_batch_pred, val_batch_labels))
        ind1 = np.where(val_batch_labels == 0); ind2 = np.where(val_batch_labels == 4)
        val_labels_1.append(val_batch_labels[ind1]); val_labels_2.append(val_batch_labels[ind2])
        val_pred_1.append(val_batch_pred[ind1]); val_pred_2.append(val_batch_pred[ind2]);
        
    acc_background = accuracy_score(val_labels_1, val_pred_1)
    acc_more_than_3 = accuracy_score(val_labels_2, val_pred_2)
    val_acc = np.mean(acc); val_acc_background = np.mean(acc_background)
    
    val_summary = tf.Summary()
    val_summary.value.add(tag="length acc", simple_value= val_acc)
    val_summary.value.add(tag="background acc", simple_value= val_acc_background)
    val_summary.value.add(tag="more than 3 acc", simple_value= acc_more_than_3)
    validate_writer.add_summary(val_summary, step)
    print('Validate acc at step %d: %f' % (step, val_acc))   
```

```python
def train_model(num_start, num_end, batch_size, model_ind, model_path, new_model_path, loadModel=False, lenght_model = False):
    
    path_to_model_pb = os.path.join( "Models", "Model" + model_ind)
    LOGDIR = "/temp/SVHN/"+model_ind
    if lenght_model:
        LOGDIR += "/len/"
    
    with tf.Session(graph=graph) as session:
        if loadModel:
            reader = tf.train.NewCheckpointReader(model_path)
            reader.get_variable_to_shape_map()
            saver.restore(session, model_path)
            print('Model restored.')
        else:
            tf.global_variables_initializer().run()
            print('Initialized.')

        train_writer = tf.summary.FileWriter(LOGDIR+"/train", session.graph)
        validate_writer = tf.summary.FileWriter(LOGDIR+"/val", session.graph)


        for step in range(num_start, num_end):
            offset = (step * batch_size) % (train_data.shape[0] - batch_size)
            batch_data = train_data[offset:(offset + batch_size), :, :, :]
            if lenght_model:
                batch_len_labels = train_length_labels[offset:(offset + batch_size)][:,:5]
                batch_labels = np.array([to_categorical(0, num_classes=11)]*batch_size*3).reshape((-1,3,11))
            else:
                batch_len_labels = np.array([to_categorical(0, num_classes=5)]*batch_size).reshape((-1,5))
                batch_labels = train_labels[offset:(offset + batch_size)]
            feed_dict = {tf_train_dataset : batch_data, tf_train_labels : batch_labels, 
                        tf_train_length: batch_len_labels}
            if lenght_model:
                summary, _, l = session.run([merge, length_optimizer, length_loss], feed_dict=feed_dict)
            else:    
                summary, _, l = session.run([merge, optimizer, logit_loss], feed_dict=feed_dict)

            # Save tensorflowboard logs
            if (step % 500 == 0):
                train_writer.add_summary(summary, step)

            #Print metric
            if (step % 200 == 0):
                print(datetime.datetime.now(),'| Minibatch loss at step %d: %f' % (step, l))


            #Save tensorflowboard logs and model
            if ((step % 2000 == 0) and (step is not 0)):
                save_path = saver.save(session, new_model_path)
                print("Model saved.")
                if lenght_model:
                    eval_length_valid_data(step, 500, validate_writer)
                else:
                    eval_valid_data(step, 500, validate_writer)


        train_writer.close()
        validate_writer.close()
        tf.train.write_graph(graph, path_to_model_pb,\
                         'Classification'+model_ind+'.pb', as_text=False)
```

## 4) Results

### Validation dataset

```python
def generate_dataset(dataset, test = False, image_size = 64, grey = False):
    "test - true when generate validate/test dataser"
    
    if grey:
        num_channels = 1
    else:
        num_channels = 3
    w = np.array([[[ 0.07, 0.72,  0.21]]])
    pixel_depth = 255.0  # Number of levels per pixel.   
    images = np.ndarray(shape=(len(dataset), image_size, image_size, num_channels),
                         dtype=np.float32)
    labels = np.ndarray(shape=(len(dataset), 3, 11))
    length_labels = np.ndarray(shape=(len(dataset), 4)) 
    num_image = 0
    
    
    for data in dataset:
        
        # Skip images with more then 5 digits
        if len(data["boxes"]) > 3:
            print("yra")
            continue
        
        # Read image
        image_path = data['filename']
        image_data = Image.open(image_path)
        
        # Get digits bounding boxes and labels
        top = []; left = []; height = []; width = []; label = []
        for box in data["boxes"]:
            top.append(box['top'])
            left.append(box['left'])
            height.append(box['height'])
            width.append(box['width'])
            if box["label"] == 10:
                label.append(0.0)
            else:
                label.append(box["label"])
        
        # Isert blank category      
        i = 3 - len(label)
        if i > 0 :
            label.extend([10.0]*i)
        #labels.append(to_categorical(label,num_classes=11))
        labels[num_image, :, :] =  to_categorical(label,num_classes=11)     
        # Get whole number bounding box
        num_top = np.amin(top)
        num_left = np.amin(left)
        num_height = np.amax(top) + height[np.argmax(top)] - num_top
        num_width = np.amax(left) + width[np.argmax(left)] - num_left  
        
        if test is True:
            image_data = image_data.crop((num_left, num_top,num_left + num_width, num_top + num_height)\
                                    ).resize([image_size,image_size])#.convert('L')
        else:
            # Expanding image by 30%
            num_bottom = np.amin([np.ceil(num_top + 1.3 * num_height), image_data.size[1]])
            num_right = np.amin([np.ceil(num_left + 1.3 * num_width), image_data.size[0]])
            num_top = np.amax([np.floor(num_top - 0.3 * num_height), 0])
            num_left = np.amax([np.floor(num_left - 0.3 * num_width), 0])         

            # Cropping the expanded bounding box 
            image_data = image_data.crop((int(num_left), int(num_top), int(num_right), int(num_bottom))\
                                        ).resize([image_size,image_size])#.convert('L')
        #Convert image to the grey scale and then normalize it
        normalized_image = (np.array(image_data) - pixel_depth / 2)/ pixel_depth
            
        images[num_image, :, :,:] = normalized_image
        length_labels[num_image, :] = to_categorical(len(data["boxes"]),num_classes=4)
        num_image+=1

    return images, labels, length_labels
```

```python
#get number from CNN outputs
def get_number(ind, data, arg_max = True):

    num = ""
    empty = False
    for i in range(3):
        if arg_max is True:
            digit = argmax(data[i])
        else:
            digit = data[i]
        if int(digit) is not 10:
            num += str(digit)
            if empty is True:
                print("Number ind:%d predicted %d is epmty and %d is not" % (ind, i,i+1))
        else:
            empty = True
        if i is 2:
            empty = False
    return num
```

```python
def get_number_with_len(length, ind, data, arg_max = True):

    num = ""
    if length > 3:
        length = 3
    for i in range(length):
        if arg_max is True:
            digit = argmax(data[i])
        else:
            digit = data[i]
        if int(digit) is not 10:
            num += str(digit)
    return num
```

```python
def get_results(data, labels, batch_size, model_path, length_model = False):
    
    curr_error_data,  curr_len_error_data, acc1, acc2, acc3, dig_acc, acc_len, acc_background  \
        = [], [], [], [], [], [], [], []
    steps = len(data) // batch_size
    
    with tf.Session(graph=graph) as session:
        reader = tf.train.NewCheckpointReader(model_path)
        reader.get_variable_to_shape_map()
        saver.restore(session, model_path)
        print('Model restored.')
        
        for j in range(steps):
            offset = (j * batch_size) % (len(data) - batch_size)
            batch_data = data[offset:(offset + batch_size), :, :, :]
            batch_labels = labels[offset:(offset + batch_size), :]
            
            if length_model: 
                batch_pred = validate_len_pred.eval(feed_dict={tf_validate_dataset : batch_data})
                acc_len.append(accuracy_score(batch_pred, argmax(batch_labels, axis = 1)))
                ind = np.where(batch_labels[:,0] == 1 )
                if len(batch_pred[ind]) is not 0:
                    acc_background.append(accuracy_score(batch_pred[ind], argmax(batch_labels[ind], axis = 1)))
            else: 
                batch_pred = validate_prediction.eval(feed_dict={tf_validate_dataset : batch_data})
                for k in range(len(batch_data)):
                    actual_number = get_number(j*batch_size+k, batch_labels[k])
                    pred_number = get_number(j*batch_size+k, batch_pred[:,k], arg_max = False)
                    if actual_number != pred_number:
                        curr_error_data.append((j*batch_size+k, pred_number, actual_number))
                        
                acc1.append(calculate_digit_acc(batch_pred, batch_labels, 0))
                acc2.append(calculate_digit_acc(batch_pred, batch_labels, 1))
                acc3.append(calculate_digit_acc(batch_pred, batch_labels, 2)) 
                dig_acc.append(np.mean([acc1,acc2,acc3]))
            

    if length_model:
        return  np.mean(acc_len), np.mean(acc_background)
    return  np.mean(acc1), np.mean(acc2), np.mean(acc3),  np.mean(dig_acc), 1 - len(curr_error_data)/len(test_data), curr_error_data
```

Remove dropout in model graph (Tflite don't support dropout op).

```python
def findDropoutOp(graphNode):
    dropString = "drop"
    dropInd = []
    dropoutOp = False
    
    for i in range(len(graphNode)):
        
        if dropoutOp and  not (dropString in graphNode[i].name):
            dropoutOp = False
        
        if (not dropoutOp) and (dropString in graphNode[i].name):
            dropInd.append(i)
            dropoutOp = True
            
    return dropInd
        
def findNodeInOp(graphNode, nodeName, startInd):
    ind = []
    for i in range(startInd, len(graphNode)):
        for j in range(len(graphNode[i].input)):
            if graphNode[i].input[j] == nodeName:
                ind.append([i,j])
    return ind

def removeDropoutOp(graphNode, dropOpLen = 12):
    #dropOpLen -dropout operations number in graph
    dropInd = findDropoutOp(graphNode)
    print(dropInd)
    indLen =len(dropInd)
                
    for i in range(indLen):
        #replace dropout variables in not dropout op nodes
        ind = findNodeInOp(graphNode, graphNode[dropInd[i]+11].name, dropInd[i]+12)
        for j in ind:
            graphNode[j[0]].input[j[1]] = graphNode[dropInd[i]-1].name
        #join new graph nodes without dropout op
        if i is 0 :
            newGraph = graphNode[:dropInd[i]]
        elif i is indLen-1:
            newGraph += graphNode[dropInd[i-1]+dropOpLen: dropInd[i]] + graphNode[dropInd[i]+dropOpLen:]
        else: 
            newGraph += graphNode[dropInd[i-1]+dropOpLen: dropInd[i]] 
            
    return newGraph 

def display_nodes(nodes):
    for i, node in enumerate(nodes):
        print('%d %s %s' % (i, node.name, node.op))
        [print(u' %d  %s' % (i, n)) for i, n in enumerate(node.input)]
```

```python
from tensorflow.core.framework import graph_pb2
frozen_model_path = os.path.join( "Models","Model_101","frozen.pb")
frozen_model_without_op_path = os.path.join( "Models","Model_101","frozen_model_without_dropout.pb")

graph = tf.GraphDef()
with tf.gfile.Open(frozen_model_path, 'rb') as f:
    data = f.read()
    graph.ParseFromString(data)
    
    newNodes = removeDropoutOp(graph.node)

    # Save graph
    output_graph = graph_pb2.GraphDef()
    output_graph.node.extend(newNodes)
    with tf.gfile.GFile(frozen_model_without_op_path, 'w') as f:
        f.write(output_graph.SerializeToString())
```
